```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.impute import MissingIndicator
%matplotlib inline 
```

```python
data = pd.read_csv('/content/uk_universities.csv')
print(data)
```

# Задача №7

```python
data.isnull().sum()
```

```python
data["Minimum_IELTS_score"] = SimpleImputer(strategy = "median").fit_transform(data[["Minimum_IELTS_score"]])

display(data[["Minimum_IELTS_score"]].isnull().sum()) 
```

Выполнена импьютация медианным значением столбца Minimum_IELTS_score.  Число нулевых значений стало 0.

# Задача №27

```python
import matplotlib.pyplot as plt 
import seaborn as sns
plt.hist(data[["Minimum_IELTS_score"]], 100) 
plt.show()
sns.violinplot(x=data["Minimum_IELTS_score"]);
```

```python
# вычисление 5% и 95% квантилей для столбца 'A'
q05 = data['Minimum_IELTS_score'].quantile(0.05)
q95 = data['Minimum_IELTS_score'].quantile(0.95)
print('\n5% квантиль: ', q05)
print('95% квантиль: ', q95)
```

```python
data.loc[data['Minimum_IELTS_score'] < q05, 'Minimum_IELTS_score'] = q05
data.loc[data['Minimum_IELTS_score'] > q95, 'Minimum_IELTS_score'] = q95
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
plt.hist(data[["Minimum_IELTS_score"]], 100) 
plt.show()
sns.violinplot(x=data["Minimum_IELTS_score"]);
```
